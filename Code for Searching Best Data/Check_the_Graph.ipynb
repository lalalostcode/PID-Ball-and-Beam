```python
import pandas as pd
import matplotlib.pyplot as plt

# Nama kolom
kolom = ['milis', 'posisi_target', 'posisi_bola']

# Baca file TXT dan konversi ke DataFrame
# Ganti delimiter sesuai isi file: ',' untuk koma, '\s+' untuk spasi
df = pd.read_csv('../BALL&BEAM/ballandbeam6.txt', header=None, names=kolom, delimiter=',')

# Simpan ke format CSV
df.to_csv('../BALL&BEAM/ballandbeam6.csv', index=False)
```

```python
df.head()
```

```python
df.describe()
```

```python
plt.figure(figsize=(10, 6))
plt.plot(df['milis'], df['posisi_target'], label='Posisi Target', linestyle='--')
plt.plot(df['milis'], df['posisi_bola'], label='Posisi Bola')
plt.xlabel('Waktu (milis)')
plt.ylabel('Posisi')
plt.title('Posisi Bola dan Target terhadap Waktu')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# Dapatkan index dari baris di mana posisi_bola == 1
index_posisi_bola_1 = df.index[df['posisi_bola'] == 1].tolist()

# Tampilkan daftar index-nya
print(index_posisi_bola_1)
```

```python
# Hapus 132 baris pertama berdasarkan posisi
df = df.iloc[1277:].reset_index(drop=True)

# Verifikasi
print(df.head())
```

```python
# Geser waktu agar mulai dari 0
milis_awal = df['milis'].iloc[0]
df['milis'] = df['milis'] - milis_awal

# Verifikasi
print(df.head())
```

```python
plt.figure(figsize=(10, 6))
plt.plot(df['milis'], df['posisi_target'], label='Posisi Target', linestyle='--')
plt.plot(df['milis'], df['posisi_bola'], label='Posisi Bola')
plt.xlabel('Waktu (milis)')
plt.ylabel('Posisi')
plt.title('Posisi Bola dan Target terhadap Waktu')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# Dapatkan index dari baris di mana posisi_bola == 1
index_posisi_bola_1 = df.index[df['posisi_bola'] == 16].tolist()

# Tampilkan daftar index-nya
print(index_posisi_bola_1)
```
